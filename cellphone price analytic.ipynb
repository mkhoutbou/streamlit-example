```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Charger le fichier CSV dans un DataFrame
data = pd.read_csv("Cellphone.csv")

# Afficher le nombre de lignes et de colonnes
print("Nombre de lignes :", data.shape[0])
print("Nombre de colonnes :", data.shape[1])

# Afficher les premires lignes du DataFrame
print(data.head())

# Vrifier le type et les statistiques de chaque variable
print(data.info())
print(data.describe())

# Vrifier la prsence de valeurs manquantes
print(data.isnull().sum())

# Vrifier la prsence de valeurs aberrantes
colonnes_numeriques = ['Price', 'Sale', 'weight', 'resoloution', 'ppi', 'cpu core', 'cpu freq', 'internal mem', 'ram', 'RearCam', 'Front_Cam', 'battery', 'thickness']
for colonne in colonnes_numeriques:
    plt.figure()
    data[colonne].plot.box()
    plt.title(colonne)

# Calculer la corrlation entre les variables
correlation = data.corr()
plt.figure(figsize=(12, 8))
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.title("Matrice de corrlation")

# Visualiser la distribution des variables
data.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```
